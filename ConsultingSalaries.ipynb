```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('202201_ConsultingSalary.xlsx',sheet_name='Sheet2')
df.tail()
```

```python
df = df.replace(r'-', np.nan, regex=True)
```

```python
sns.boxplot(data=df, x="Position", y = "Salary")
plt.title('MBB Base Salary by position (Glassdoor 01/2022)')
```

```python
sns.boxplot(data=df, x="Position", y = "Salary", hue = 'Company')
plt.title('MBB Base Salary in different countries (Glassdoor 01/2022)')
```

```python
fig, ax = plt.subplots(figsize =(10, 6))
sns.boxplot(data=df, hue="Position", y = "Salary", x = 'Country')
plt.title('MBB Base Salary in different countries (Glassdoor 01/2022)')
plt.xticks(rotation=30, ha = 'right');
```
